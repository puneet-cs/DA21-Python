## Match case statement

```python
# Match case statement is used to replace multiple if else conditions

# Example-1
num = int(input("Enter a Number"))

match num:
    case 10:
        print("TEN")
    case 20:
        print("TWENTY")
    case 30:
        print("Thirty")
    case 40:
        print("forty")
    case 50:
        print("fifty")
    case _:
        print("Something else")
```

```python
# Interview Question
x = 15
if x > 3:
    print("A")
    if x%3 == 0 and x:
        print("B")
    elif x == 15:
        print("D")
    else:
        pass
else:
    print("E")
```

```python
# Example-7: nested loop
# for every value of i, internal loop will completely execute

i = 1
while i<= 3:  // outer
    j = 1
    while j<=2:   // inner
        print(i,j)
        j+=1
    i+=1
```

## Inserting items inside the list

```python
# append(), extend(), insert()
```

```python
l3 = []

# append() insert the item at the last in the list . O(1)
l3.append(100)
print(l3)
```

```python
# extend() is used to insert many elements together. O(k)
l3.extend([200, 300, 400])
print(l3)
```

```python
# insert() function is used to insert element at a particular index. O(n)
l3.insert(1, "Hello")
print(l3)
```
