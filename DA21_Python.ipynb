## Match case statement

```python
# Match case statement is used to replace multiple if else conditions

# Example-1
num = int(input("Enter a Number"))

match num:
    case 10:
        print("TEN")
    case 20:
        print("TWENTY")
    case 30:
        print("Thirty")
    case 40:
        print("forty")
    case 50:
        print("fifty")
    case _:
        print("Something else")
```

```python
# Interview Question
x = 15
if x > 3:
    print("A")
    if x%3 == 0 and x:
        print("B")
    elif x == 15:
        print("D")
    else:
        pass
else:
    print("E")
```

```python
# discard() : delete the element by its value
lunch_set.discard('Lemon rice')
print(lunch_set)
```

```python
lunch_set.remove("5 Meals")
print(lunch_set)
```

```python
lunch_set.discard("5 Meals")
print(lunch_set)
```

```python
# Remove() function will show the "error" if item don't present in the set.
# discard() function will NOT raise any "error" even though the element is not present in the set.
```

```python
lunch_set.pop()
print(lunch_set)
```

```python
# clear(): delete all the items
lunch_set.clear()
print(lunch_set)
```

```python
# Interview Question
def fun(d):
    total = 0;
    for value in d.values():
        if type(value) == dict:
            total += fun(value)
        else:
            total += value
    return total

d = {'a' : 1, 'b' : {'c' : 2, 'd' : 3}, 'e' : 4}
print(fun(d))
```

## Default parameters

```python
# The default parameter value of GF1 here is "Pooja"
def Greesh(GF1 = "Pooja"):
    print(GF1)

Greesh("Kajal")
Greesh()
```

```python
def Greesh(GF1, GF2 = "Pooja"):
    print(GF1, " : ", GF2)

Greesh("Kajal")
Greesh("Kajal", "Samantha")
```

## Passing arbitrary number of arguments using *args

```python
def Greesh(*GF):
    for i in GF:
        print(i)

Greesh()   # 0 argument
# Greesh("Kajal")
# Greesh("Kajal", "Pooja")
Greesh("Kajal", "Pooja", "Samantha")
```

```python
# create a function which can receive any number of intergers and return the sum of all the values

def add(*val):
    total = 0
    for i in val:
        total += i
    return total

print(add(10,30))
print(add(10,30, 40, 50, 60, 10, 30, 40, 50))
```

```python
def add(*val):
    return sum(val)

print(add(10,30))
print(add(10,30, 40, 50, 60, 10, 30, 40, 50))
```
