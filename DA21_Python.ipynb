# Type casting

```python
v1 = int(20.2)  # float to int
v2 = int("300")  # string to int
# int function cannot convert string of characters and string of float into integer.

v3 = int(False)  # Bool to int
v4 = int("1010", 2)
v5 = int("C", 16)   # 0-9, A-F
```

```python
print(type(v1), v1)
print(type(v2), v2)
print(type(v3), v3)
print(type(v4), v4)
print(type(v5), v5)
```

# Logical Operators

```python
# and, or, not

a = True
b = False
c = True

print(a and b)
print(a and c)
print(b and b)

# and : if both are True then output is True
# if any of the input is false output is False
```

```python
a = True
b = False
c = True

print(a or b)  # True
print(a or c)  # True
print(b or b)  # False

# or : if any of the input is True, output is True
```

```python
print(not a)
```

```python
print(not b)
```

```python
a = 10
print(a > 10 or a <= 10)
```

```python
print(not(a > 10) and a <= 10)
```
